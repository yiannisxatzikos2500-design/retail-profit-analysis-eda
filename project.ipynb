```python
# 1. Standardize Column Names 
df.columns = [col.replace(' ', '_').replace('-', '_').lower() for col in df.columns]

# 2. Fix Date Formats
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

# 3. Verify Data Integrity
print(f" Step 1 Complete: {df.shape[0]} rows loaded.")
print(f"Missing Values Found: {df.isnull().sum().sum()}") 

# Show the cleaned data
df.head()
```

```python
# 1. Calculate Shipping Delay (Efficiency)
# This tells us how many days it took to fulfill the order
df['ship_days'] = (df['ship_date'] - df['order_date']).dt.days

# 2. Calculate Profit Margin
# This shows the percentage of Sales that is actually Profit
# Margin = Profit / Sales
df['profit_margin'] = df['profit'] / df['sales']

# 3. Verify the new columns
print("--- NEW BUSINESS METRICS CREATED ---")
display(df[['order_date', 'ship_date', 'ship_days', 'profit', 'sales', 'profit_margin']].head())
```

```python
# 1. Total Cost (Sales minus what we kept)
df['total_cost'] = df['sales'] - df['profit']

# 2. Unit Cost 
df['unit_cost'] = df['total_cost'] / df['quantity']

# 3. Markup Multiplier (How much we upcharged the cost)
df['markup_ratio'] = df['sales'] / df['total_cost']

print("--- REVENUE VS COST METRICS ---")
display(df[['sales', 'total_cost', 'profit', 'unit_cost', 'markup_ratio']].head())
```

Clustering 

We are using the elbow method to find out the number of clusters we need to make

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# SELECT FEATURES (Using Sales and Profit for the Elbow)
X = df[['sales', 'profit']].dropna()

# SCALE THE DATA (K-Means is distance-based, scaling is mandatory)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# RUN THE ELBOW CALCULATION
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# PLOT THE ELBOW GRAPH
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o', linestyle='--', color='b')

plt.title('The Elbow Method: Finding the Optimal Clusters', fontsize=15)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Sum of Squared Distances)', fontsize=12)
plt.xticks(K_range)
plt.grid(True, alpha=0.3)
plt.show()

```

So we can see that 3 clusters is the ideal number

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. PREPARE THE DATA
cust_data = df.groupby('customer_id').agg({
    'sales': 'sum',
    'profit': 'sum',
    'quantity': 'sum',
    'order_id': 'count'
}).rename(columns={'order_id': 'frequency'})

# 2. SCALE AND CLUSTER (3 GROUPS)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(cust_data)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cust_data['cluster'] = kmeans.fit_predict(scaled_features)

# 3. LABEL BY PROFIT
rank = cust_data.groupby('cluster')['profit'].mean().sort_values(ascending=False).index
mapping = {rank[0]: 'High Value (Whales)', rank[1]: 'Mid Value', rank[2]: 'Low Value / At-Risk'}
cust_data['segment'] = cust_data['cluster'].map(mapping)

# 4. THE DOT GRAPH (SCATTER PLOT)
plt.figure(figsize=(12, 8))

# Define colors for the dots
palette = {'High Value (Whales)': '#2ecc71', 'Mid Value': '#3498db', 'Low Value / At-Risk': '#e74c3c'}

sns.scatterplot(
    data=cust_data, 
    x='sales', 
    y='profit', 
    hue='segment', 
    palette=palette, 
    s=100, 
    alpha=0.7, 
    edgecolor='black'
)

# Formatting the graph
plt.title('Customer Clusters: Sales vs. Profit', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Total Profit ($)', fontsize=12)
plt.axhline(0, color='black', linestyle='--', alpha=0.5) # Shows the break-even line
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(title='Customer Segment', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

total_profit_df = cust_data.groupby('segment')['profit'].sum().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(data=total_profit_df, x='segment', y='profit', palette=palette, hue='segment', legend=False)
plt.title('Total Profit per Segment (Summary)', fontsize=14)
plt.show()
```

By comparing the high-value group with the low-value segment, we can determine whether specific customers are exploiting our pricing or if internal errors are causing the profit loss in this section. This comparison allows us to identify if the deficit is driven by customer behavior or a flaw in our own sales strategy.

```python
# 1. Get the IDs of the Low Value customers
low_value_ids = cust_data[cust_data['segment'] == 'Low Value / At-Risk'].index

# 2. Pull their original transaction data 
low_value_transactions = df[df['customer_id'].isin(low_value_ids)]
high_value_transactions = df[df['customer_id'].isin(cust_data[cust_data['segment'] == 'High Value (Whales)'].index)]

# 3. Compare 
stats_comparison = pd.DataFrame({
    'Metric': ['Avg Discount', 'Avg Shipping Cost', 'Profit per Order', 'Sales per Order'],
    'Low Value Group': [
        low_value_transactions['discount'].mean(),
        low_value_transactions['shipping_cost'].mean() if 'shipping_cost' in df.columns else 0,
        low_value_transactions['profit'].mean(),
        low_value_transactions['sales'].mean()
    ],
    'Whales Group': [
        high_value_transactions['discount'].mean(),
        high_value_transactions['shipping_cost'].mean() if 'shipping_cost' in df.columns else 0,
        high_value_transactions['profit'].mean(),
        high_value_transactions['sales'].mean()
    ]
})

print("--- WHY ARE THEY LOW VALUE? (Comparison Stats) ---")
print(stats_comparison)
```

The average discount is nearly identical across both groups. This confirms that the profit loss in the low-value segment isn't caused by deeper discounts

```python
# 1. Prep data for both segments
low_ids = cust_data[cust_data['segment'] == 'Low Value / At-Risk'].index
high_ids = cust_data[cust_data['segment'] == 'High Value (Whales)'].index

# 2. Aggregated stats by Sub-Category
def get_basket(ids):
    return df[df['customer_id'].isin(ids)].groupby('sub_category').agg({
        'profit': 'sum',
        'discount': 'mean',
        'sales': 'sum'
    })

low_basket = get_basket(low_ids)
high_basket = get_basket(high_ids)

# 3. Join them for a direct comparison
comparison = low_basket.join(high_basket, lsuffix='_Low', rsuffix='_High').fillna(0)

# 4. Plot Profit Comparison
plt.figure(figsize=(14, 8))
comparison[['profit_Low', 'profit_High']].sort_values('profit_Low').plot(
    kind='barh', 
    color=['#e74c3c', '#2ecc71'], 
    figsize=(12, 10)
)
plt.title('Profit Comparison: Low Value vs. Whales by Product', fontsize=16)
plt.axvline(0, color='black', lw=1)
plt.xlabel('Total Profit ($)')
plt.show()
```

We have identified a significant loss in certain machine sales. This disparity is particularly evident when comparing the high-value segment, which remains profitable

```python
# 1. Get the transactions for both groups
low_value_ids = cust_data[cust_data['segment'] == 'Low Value / At-Risk'].index
high_value_ids = cust_data[cust_data['segment'] == 'High Value (Whales)'].index

low_val_ship = df[df['customer_id'].isin(low_value_ids)]['ship_mode'].value_counts(normalize=True)
high_val_ship = df[df['customer_id'].isin(high_value_ids)]['ship_mode'].value_counts(normalize=True)

# 2. Combine for comparison
ship_comparison = pd.DataFrame({
    'Low Value %': low_val_ship,
    'Whales %': high_val_ship
}).fillna(0)

print("--- SHIPPING MODE PREFERENCE ---")
print(ship_comparison)

# 3. Visualize 
ship_comparison.plot(kind='bar', figsize=(10, 6), color=['#e74c3c', '#2ecc71'])
plt.title('Delivery Class: Low Value vs. Whales', fontsize=14)
plt.ylabel('Percentage of Orders')
plt.xlabel('Shipping Mode')
plt.xticks(rotation=45)
plt.show()
```

The delivery class is also nearly identical between the two groups. This indicates that the profit loss is not being driven by more expensive shipping methods

```python
# 1. Isolate Machine sales for both groups
machine_df = df[df['sub_category'] == 'Machines']

low_ids = cust_data[cust_data['segment'] == 'Low Value / At-Risk'].index
high_ids = cust_data[cust_data['segment'] == 'High Value (Whales)'].index

# 2. Compare the math
machine_comparison = pd.DataFrame({
    'Metric': ['Avg Sales Price', 'Avg Discount %', 'Avg Profit per Unit'],
    'Low Segment': [
        machine_df[machine_df['customer_id'].isin(low_ids)]['sales'].mean(),
        machine_df[machine_df['customer_id'].isin(low_ids)]['discount'].mean(),
        machine_df[machine_df['customer_id'].isin(low_ids)]['profit'].mean()
    ],
    'Whale Segment': [
        machine_df[machine_df['customer_id'].isin(high_ids)]['sales'].mean(),
        machine_df[machine_df['customer_id'].isin(high_ids)]['discount'].mean(),
        machine_df[machine_df['customer_id'].isin(high_ids)]['profit'].mean()
    ]
})

print("--- THE MACHINE PROFIT GAP ---")
print(machine_comparison)
```

It also helps identify where we are losing profit; for example, analyzing the low-value segment reveals they were classified that way because they primarily buy items we sell at a loss, specifically tables and bookcases.
